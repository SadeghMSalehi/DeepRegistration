```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import pandas as pd
```

```python
results = pd.read_csv('../../results/3d_3d_4.csv')
separat_subject = {}
```

```python
for i in range(27, 37):
    df = results[results['Subject Name'].str.contains(str(i)+'_')]
    separat_subject[str(i)]=df['Rot distance']
```

```python
box = plt.boxplot([separat_subject['27'],
                   separat_subject['28'],
                   separat_subject['29'],
                   separat_subject['30'],
                   separat_subject['31'],
                   separat_subject['32'],
                   separat_subject['33'],
                   separat_subject['34'],
                   separat_subject['35'],
                   separat_subject['36']],
                   notch=True, patch_artist=True)
plt.ylim(0, 20)
plt.savefig('../../results/3d_3d-27-36_4.eps', format='eps', dpi=1000)
plt.show()
```
